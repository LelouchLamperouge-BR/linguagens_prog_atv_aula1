## 4. Batalha 1 — A tirania dos tipos de dados

Nos slides, vimos que datas não podem ser tratadas como texto e que números em formato string quebram análises. fileciteturn2file0

### Tarefa
Converta corretamente, quando fizer sentido:
- `data`
- `receita`
- `lucro`
- `quantidade`

### Orientação
- Investigue valores estranhos antes de converter
- Pense no uso de `errors='coerce'`
- Registre em markdown ou comentário quais problemas encontrou


```python
# Explore valores problemáticos e faça as conversões de tipo

df['receita'].values
df['receita'].dtype
df['receita'].isna().sum()
#o dtype da coluna é object e os valores estão mixed (float e string) e não padronizados (alguns valores estão cifrados).

df['receita'] = pd.to_numeric(df['receita'], errors='coerce')
#convertendo coluna para tipo númerico e setando erros de parsing para valores NaN

df['lucro'].values
df['lucro'].dtype
df['lucro'].isna().sum()
#lucro tem NaN porém todos os valores válidos são do tipo correto (float). Logo, se tratando de números não há necessidade de conversão

df['quantidade'].values
df['quantidade'].dtype
df['quantidade'].isna().sum()
#coluna quantidade está perfeitamente padronizada. não há o que ser feito

df['data'].values
df['data'].dtype
df['data'].isna().sum()
#coluna data está com o tipo Objeto E não existe um formato padrão de data, existem vários formatos de data no dataset, logo há a necessidade de padronização dessa informação

df['data'] = pd.to_datetime(df['data'], errors='coerce')
#convertendo coluna para tipo data

#o parser de pd.to_datetime() não conseguiu inferir bem os diferentes formatos
#de data, sendo necessário uma lógica mais complexa e customizada para
#normalizar os dados
```

###R: a mesma resposta vale para ambas perguntas. Tratar datas e números como strings é um problema pois strings são dados literais, logo não são apropriadas para manipulações algébricas e/ou tratar dados serializados pois strings não possuem magnitude. Realizamos operações matemáticas com números e, por mais que não pareça, datetime também envolve manipulação matemática pois a contagem de tempo é feita através de timestamp, logo uma string é um tipo de dado ineficaz para estes dois casos
